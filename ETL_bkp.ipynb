# Create new dataframe with select columns and renaming the columns

```python
# Diversity Index Dataframe

# Rename the column names of the Diversity Index Dataframe
new_diversityindex_df = diversityindex_df.rename(columns = {"Diversity-Index": "Diversity Index","Black or African American alone, percent, 2013":"Black/African American",
                                                           "American Indian and Alaska Native alone, percent, 2013":"American Indian & Alaska Native",
                                                           "Asian alone, percent, 2013": "Asian","Native Hawaiian and Other Pacific Islander alone, percent,":
                                                           "Native Hawaiian & Other Pacific Islander","Two or More Races, percent, 2013":"Two or More Races",
                                                           "Hispanic or Latino, percent, 2013":"Hispanic/Latino","White alone, not Hispanic or Latino, percent, 2013"
                                                           :"White,not Hispanic/Latino"})



# Splitting the Location Column to two columns (State and County)
new = new_diversityindex_df["Location"].str.split(",", n = 0, expand = True)
new_diversityindex_df["State"]= new[1]
new_diversityindex_df["County"]= new[0]

#Move the State and County Columns to the first and second position
cols =new_diversityindex_df.columns.tolist()
cols = cols[-2:] + cols[:-2]
new_diversityindex_df = new_diversityindex_df[cols]

#Drop the Location Column
final_diversityindex_df = new_diversityindex_df.drop(["Location"],axis = 1)

final_diversityindex_df.head()
```

```python
# Unemployment Rate and Median Household Income Dataframe

# Creating new dataframe by Dropping the columns which are not needed.We need only 2013 dta.
new_unemploymentrate_df = unemployment_df.drop(['Rural_urban_continuum_code_2013',
       'Urban_influence_code_2013', 'Metro_2015', 'Civilian_labor_force_2007',
       'Employed_2007', 'Unemployed_2007', 'Unemployment_rate_2007',
       'Civilian_labor_force_2008', 'Employed_2008', 'Unemployed_2008',
       'Unemployment_rate_2008', 'Civilian_labor_force_2009', 'Employed_2009',
       'Unemployed_2009', 'Unemployment_rate_2009',
       'Civilian_labor_force_2010', 'Employed_2010', 'Unemployed_2010',
       'Unemployment_rate_2010', 'Civilian_labor_force_2011', 'Employed_2011',
       'Unemployed_2011', 'Unemployment_rate_2011',
       'Civilian_labor_force_2012', 'Employed_2012', 'Unemployed_2012',
       'Unemployment_rate_2012','Civilian_labor_force_2014', 'Employed_2014', 'Unemployed_2014',
       'Unemployment_rate_2014', 'Civilian_labor_force_2015', 'Employed_2015',
       'Unemployed_2015', 'Unemployment_rate_2015','Civilian_labor_force_2016', 'Employed_2016', 'Unemployed_2016',
       'Unemployment_rate_2016','Civilian_labor_force_2013','Employed_2013','Unemployed_2013'], axis = 1)

new_unemploymentrate_df.head()
```

```python
# Rename the Columns of the Dataframe
final_unemployment_df = new_unemploymentrate_df.rename(columns = {"Area_name":"County","Unemployment_rate_2013":"Unemployment_rate","Med_Household_Income_2015"
                                                           :"Med_Household_Income","Med_HH_Income_Percent_of_State_Total_2015":
                                                           "Med_HH_Income_Percent"})

# Remove the State abbreviation from the County/Area_name Column
new = final_unemployment_df["County"].str.split(",", n = 0, expand = True)
final_unemployment_df["County"]= new[0] 

final_unemployment_df.head()
```

```python
# Education Dataframe

# Creating new dataframe by Dropping the columns which are not needed. We need only 2013 data.
new_education_df = education_df.drop(['2003 Rural-urban Continuum Code',
       '2003 Urban Influence Code', '2013 Rural-urban Continuum Code',
       '2013 Urban Influence Code', 'Less than a high school diploma, 1970',
       'High school diploma only, 1970', 'Some college (1-3 years), 1970',
       'Four years of college or higher, 1970',
       'Percent of adults with less than a high school diploma, 1970',
       'Percent of adults with a high school diploma only, 1970',
       'Percent of adults completing some college (1-3 years), 1970',
       'Percent of adults completing four years of college or higher, 1970',
       'Less than a high school diploma, 1980',
       'High school diploma only, 1980', 'Some college (1-3 years), 1980',
       'Four years of college or higher, 1980',
       'Percent of adults with less than a high school diploma, 1980',
       'Percent of adults with a high school diploma only, 1980',
       'Percent of adults completing some college (1-3 years), 1980',
       'Percent of adults completing four years of college or higher, 1980',
       'Less than a high school diploma, 1990',
       'High school diploma only, 1990',"Some college or associate's degree, 1990",
       "Bachelor's degree or higher, 1990",
       "Percent of adults with less than a high school diploma, 1990",
       "Percent of adults with a high school diploma only, 1990",
       "Percent of adults completing some college or associate's degree, 1990",
       "Percent of adults with a bachelor's degree or higher, 1990",
       "Less than a high school diploma, 2000",
       "High school diploma only, 2000",
       "Some college or associate's degree, 2000",
       "Bachelor's degree or higher, 2000",
       "Percent of adults with less than a high school diploma, 2000",
       "Percent of adults with a high school diploma only, 2000",
       "Percent of adults completing some college or associate's degree, 2000",
       "Percent of adults with a bachelor's degree or higher, 2000",
        "Less than a high school diploma, 2011-2015",
        "High school diploma only, 2011-2015","Bachelor's degree or higher, 2011-2015",
        "Some college or associate's degree, 2011-2015"],axis = 1)
                                        
# Removing the first row from the Dataframe which has the education details of US which we don't need
new_education_df1 = new_education_df.drop(0)
new_education_df1.head()
```

```python
# Rename the column names of the education dataframe
final_education_df = new_education_df1.rename(columns = {"Area name":"County",
                                                       "Percent of adults with less than a high school diploma, 2011-2015":"Less than a high school diploma(%)",
                                                       "Percent of adults with a high school diploma only, 2011-2015":"High school diploma(%)",
                                                       "Percent of adults completing some college or associate's degree, 2011-2015":"Some college or associate's degree(%)",
                                                       "Percent of adults with a bachelor's degree or higher, 2011-2015":"Bachelor's degree or higher(%)"})

final_education_df.head()
```

# Cleaning the Dataframe

```python
# Diversity Index Dataframe

#Find null values in the Diversity Index Dataset
final_diversityindex_df.isnull().sum()

# Check the Unique Values of State Column
final_diversityindex_df['State'].unique().tolist()

# Delete the rows where the state column has null values
final_diversityindex_df1 = final_diversityindex_df.dropna(axis=0, subset=['State'])
final_diversityindex_df1

# Verify if the new dataframe has any null values
final_diversityindex_df1.isnull().sum()


#Remove all spaces in State Column
final_diversityindex_df1['State'] = final_diversityindex_df1['State'].str.strip()

# final cleaned dataframe
final_diversityindex_df1
```

```python
# Find null values in the Unemployment Rate Dataset
final_unemployment_df.isnull().sum()

# Fill the null values with 0
final_unemployment_df1 = final_unemployment_df.fillna(0)

# Verify if the dataframe has any null values
final_unemployment_df1.isnull().sum()

# final cleaned dataframe
final_unemployment_df1.head()
```

```python
# Checking for null values
final_education_df.isnull().sum()

# Fill the null values with 0
final_education_df1 = final_education_df.fillna(0)

# Verify null values
final_education_df1.isnull().sum()

# Cleaned Dataframe
final_education_df1.head()
```

# Merging Dataset 

```python
# Merging Diversity Index Dataframe and Unemployment Dataframe on State and County
merge_df = pd.merge(final_diversityindex_df1,final_unemployment_df1, on = ["State","County"])
merge_diver_unemp_df = merge_df.sort_values(by=["State"])
merge_diver_unemp_df

#Merging education dataframe with above
merge_df1 = pd.merge(merge_diver_unemp_df,final_education_df1, on = ["State","County","FIPS Code"])
merge_diver_unemp_edu = merge_df1.sort_values(by=["State"])

#Final merged Dataframe
merge_diver_unemp_edu.columns
merge_diver_unemp_edu.head()
```

# Final merged Dataframe

```python
#Renaming the final Dataframe

merge_diver_unemp_edu_df = merge_diver_unemp_edu.rename(columns = {"Diversity Index":"DiversityIndex","Black/African American":"Black",
                            "American Indian & Alaska Native":"AmericanIndian","Native Hawaiian & Other Pacific Islander":"NativeHawaiian",
                            "Two or More Races":"TwoOrMoreRaces","Hispanic/Latino":"Hispanic","White,not Hispanic/Latino":"White",
                            "Unemployment_rate":"UnemploymentRate","Med_Household_Income":"MedianHouseHoldIncome","Med_HH_Income_Percent":"MedianHHIncomePer",
                            "Less than a high school diploma(%)":"LessthanHighSchoolDip","High school diploma(%)":"HighSchoolDiploma","Some college or associate's degree(%)":
                            "CollegeDegree","Bachelor's degree or higher(%)":"BachelorsDegree"})
merge_diver_unemp_edu_df


#Code to move the FIPS Code to first column of dataframe
cols = list(merge_diver_unemp_edu_df)
# move the column to head of list using index, pop and insert
cols.insert(0, cols.pop(cols.index('FIPS Code')))
cols
# use ix to reorder
merge_diver_unemp_edu_df = merge_diver_unemp_edu_df.ix[:, cols]

merge_diver_unemp_edu_df.head()

```
